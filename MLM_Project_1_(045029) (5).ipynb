**Clustering:**


Clustering is an unsupervised learning technique used to group similar data points together in a dataset. The goal is to partition the data into clusters, where objects within the same cluster are more similar to each other than to those in other clusters. Clustering algorithms assign data points to clusters based on certain criteria, such as distance measures or density, and can be used for various tasks such as pattern recognition, data analysis, and anomaly detection.

**In this Project we have done K-mean Clustering**

K-means clustering is an unsupervised machine learning algorithm designed to partition a dataset into a predetermined number of clusters, aiming to group similar data points together while minimizing the variance within each cluster. Through an iterative process, K-means initializes cluster centroids randomly and assigns data points to the nearest centroid based on a distance metric, often the Euclidean distance. It then updates the centroids by computing the mean of the data points assigned to each cluster, repeating this process until convergence.

The primary objective of K-means clustering is to minimize the within-cluster sum of squared distances, known as WCSS or inertia. This mathematical objective function ensures that each cluster is compact and well-separated from others. Widely applied across various domains such as market segmentation, image compression, and anomaly detection, K-means clustering also holds significant potential in analyzing cardiovascular data. By grouping individuals based on demographic features, health indicators, and lifestyle factors, K-means clustering can reveal distinct risk profiles for cardiovascular diseases and facilitate personalized healthcare interventions.

**Elbow Curve**


X-axis: Values of K (number of clusters)
Y-axis: Within Cluster Sum of Squares (WCSS) distance

The WCSS represents the sum of the squared distances between each data point and its assigned cluster center (centroid). In simpler terms, it shows how spread out the data points are within each cluster. Ideally, we want to minimize the WCSS to have tight clusters where data points are close to their respective centroids.

**Interpretation:**

The graph plots the WCSS for different K values. As the number of clusters increases, the WCSS naturally decreases because each data point gets its own cluster, minimizing the distance to its centroid. However, this improvement diminishes as we continue increasing K. The “elbow” point on the curve signifies the inflection point where the WCSS decrease slows down significantly. This point is considered the optimal number of clusters (k) as it balances the trade-off between minimizing WCSS and having too many clusters (which can lead to overfitting).

The elbow appears to be around K=4 or K=5.
Suggests that 4 or 5 clusters might be a good choice for this dataset.

The silhouette score (sscore) and Davies-Bouldin score (dbscore) are evaluation metrics used to assess the quality of clustering results:

**Silhouette Score:**

The silhouette score measures how similar an object is to its own cluster compared to other clusters.

It ranges from -1 to 1, where a higher value indicates that the object is well-matched to its own cluster and poorly matched to neighboring clusters.
A silhouette score close to 1 suggests that the clusters are well-separated, while a score close to -1 indicates that data points may have been assigned to the wrong cluster.

In the provided code, higher silhouette scores (e.g., 0.959 for 2 clusters) suggest better cluster separation and cohesion.

**Davies-Bouldin Score:**

The Davies-Bouldin score quantifies the average similarity between each cluster and its most similar cluster, where lower values indicate better clustering.
It measures the ratio of the within-cluster scatter to the between-cluster separation.

A lower Davies-Bouldin score indicates that the clusters are well-separated and have high intra-cluster similarity.

Ilower Davies-Bouldin scores (e.g., 0.353 for 2 clusters) suggest better-defined clusters with greater separation.

```python
# Required Libraries
import pandas as pd, numpy as np # For Data Manipulation
import matplotlib.pyplot as plt, seaborn as sns # For Data Visualization
import scipy.cluster.hierarchy as sch # For Hierarchical Clustering
from sklearn.cluster import AgglomerativeClustering as agclus, KMeans as kmclus # For Agglomerative & K-Means Clustering
from sklearn.metrics import silhouette_score as sscore, davies_bouldin_score as dbscore # For Clustering Model Evaluation
```

```python
wcssd = [] # Within-Cluster-Sum-Squared-Distance
nr_clus = range(1,11) # Number of Clusters
for k in nr_clus:
    kmeans = kmclus(n_clusters=k, init='random', random_state=111)
    kmeans.fit(df_ppd)
    wcssd.append(kmeans.inertia_)
plt.plot(nr_clus, wcssd, marker='x')
plt.xlabel('Values of K')
plt.ylabel('Within Cluster Sum Squared Distance')
plt.title('Elbow Curve for Optimal K')
plt.show()
```

## **Gaussen Model**

```python
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import StandardScaler

# Assuming you have a DataFrame called 'df_ppd' with preprocessed data
# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_ppd)

# Initialize and fit the Gaussian Mixture Model
n_components = 2  # You can adjust the number of components/clusters
gmm = GaussianMixture(n_components=n_components, random_state=142)
gmm.fit(X_scaled)

# Get cluster labels
cluster_labels = gmm.predict(X_scaled)

# Make a copy of the DataFrame
df_ppd_gausscluster = df_ppd.copy()

# Reset the index of the DataFrame
df_ppd_gausscluster.reset_index(level=0, inplace=True)

# Add cluster labels to the DataFrame
df_ppd_gausscluster['Cluster'] = cluster_labels

# Sort the DataFrame based on the cluster labels
df_ppd_gausscluster.sort_values('Cluster', inplace=True)

# Display the DataFrame with cluster labels
df_ppd_gausscluster
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score

# Calculate silhouette score
silhouette_score_value = silhouette_score(X_scaled, cluster_labels)

# Calculate Davies-Bouldin score
davies_bouldin_score_value = davies_bouldin_score(X_scaled, cluster_labels)

print("Silhouette Score:", silhouette_score_value)
print("Davies-Bouldin Score:", davies_bouldin_score_value)
```
